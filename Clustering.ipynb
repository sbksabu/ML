```python
import pandas_ml as pdml

import pandas as pd
```

```python
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/raw_data.csv')
```

```python
mf = pdml.ModelFrame(df.to_dict())

mf.head()
```

```python
mf.shape
```

```python
mf = mf[['ID', 'cancer', 'diabetes', 'heart_disease', 'ever_smoked']]
```

```python
mf.head()
```

```python
le = mf.preprocessing.LabelEncoder()
```

```python
mf['cancer'] = le.fit_transform(mf['cancer'])

mf.head
```

```python
mf['diabetes'] = le.fit_transform(mf['diabetes'])

mf.head
```

```python
mf.describe
```

```python
mf['heart_disease'] = le.fit_transform(mf['heart_disease'])

mf['ever_smoked'] = le.fit_transform(mf['ever_smoked'])
```

```python
mf.head
```

```python
mf.describe()
```

```python
estimator = mf.cluster.KMeans(n_clusters=4)
```

```python
estimator
```

```python
mf.fit(estimator)
```

```python
clusters = mf.predict(estimator)
```

```python
clusters.head
```

```python
clusters.name = 'cluster_id'
```

```python
clustered_mf = pd.concat([mf, clusters], axis = 1)

clustered_mf.head(10)
```

```python
mf.describe()
```

```python
cluster_means = clustered_mf.groupby(['cluster_id']).mean()
```

```python
cluster_means
```

```python
cluster_means['counts'] = pd.Series(clustered_mf.groupby(['cluster_id']).size())

cluster_means
```

```python
clustered_mf[clustered_mf['cluster_id'] == 2]
```
