```python
import pandas_ml as pdml

import pandas as pd
```

```python

house_data = pd.read_csv("/content/kc_house_data.csv")
```

```python
import warnings

warnings.filterwarnings("ignore")
```

```python
house_data.head()
```

```python
house_data.columns
```

```python
x_data = house_data[['sqft_living', 'bedrooms', 'bathrooms','yr_built', 'yr_renovated', 'grade', 'waterfront', 'view' ]]

y_data = house_data['price']
```

```python
x_data.head()
```

```python
hs_zipcode = pd.get_dummies(house_data['zipcode'])

hs_zipcode.head()
```

```python
hs_zipcode.shape
```

```python
data = pd.concat([x_data, hs_zipcode, y_data], axis=1)
```

```python
data.head()
```

```python
mf = pdml.ModelFrame(data.to_dict())
```

```python
mf.head()
```

```python
mf.target_name = 'price'
```
